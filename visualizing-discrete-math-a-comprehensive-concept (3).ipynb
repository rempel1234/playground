### 2.1 Sets and Set Theory

Sets and set theory are fundamental components of discrete mathematics (Harvey & Rodger, 2005). They encompass key operations such as unions, intersections, and differences. Examples of these operations include:

- **Unions**: Combining all elements from two or more sets, e.g., $( A \cup B )$.
- **Intersections**: Finding common elements between sets, e.g., $( A \cap B )$.
- **Differences**: Identifying elements in one set but not in another, e.g., $( A - B )$.

These operations are crucial for understanding more complex mathematical concepts such as:

- **Functions and Relations**: How sets interact with each other through mappings and relations.
- **Algorithms and Complexity**: Applications in sorting, searching, and optimization problems.
- **Probability and Statistics**: Used in calculating probabilities and analyzing data.

Additionally, set theory underpins various applications and theories in discrete mathematics, including:

- **Database Design**: Using set operations to query and manipulate data.
- **Graph Theory**: Representing and analyzing networks and relationships.
- **Combinatorics**: Counting and arranging elements in sets.

#### 2.1.1.1 Introduction to Derivation and Code

In this section, we explore how set theory is applied to real-world data through an example of finding common customers across two datasets. The derivation below demonstrates the use of set intersection to identify customers who appear in both a retail customer database and a loyalty program database. By calculating the intersection of these two sets, we can determine which customers are present in both datasets. Following the theoretical derivation, the Python code provided illustrates how to implement this concept programmatically, showcasing how set operations can be used to handle and analyze real-world data effectively.

#### 2.1.1.2 Derivation: Finding Common Customers in Retail and Loyalty Databases

Given two datasets:
- **Retail Customers Dataset:** $R$ consisting of records $\{r_1, r_2, r_3, r_4, r_5\}$
- **Loyalty Program Dataset:** $L$ consisting of records $\{l_1, l_2, l_3, l_4, l_5\}$

Where:
$$
\begin{aligned}
R = \{(101, \text{John Smith}), (102, \text{Jane Doe}), (103, \text{Emily Johnson}), (104, \text{Michael Brown}), (105, \text{Sarah Davis})\} \\
L = \{(104, \text{Michael Brown}), (105, \text{Sarah Davis}), (106, \text{Laura Wilson}), (107, \text{James Taylor}), (108, \text{Olivia Anderson})\}
\end{aligned}
$$

To find the common customers in both datasets, we compute the intersection of sets $R$ and $L$. The intersection of two sets $A$ and $B$ is given by:

$A \cap B = \{ x \mid x \in A \text{ and } x \in B \}$

Applying this to our datasets:

$R \cap L = \{ r \mid r \in R \text{ and } r \in L \}$

Substituting the values:

$R \cap L = \{ (104, \text{Michael Brown}), (105, \text{Sarah Davis}) \}$

Thus, the common customers who appear in both the retail and loyalty program datasets are:

$\{ (104, \text{Michael Brown}), (105, \text{Sarah Davis}) \}$

This result shows that Michael Brown and Sarah Davis are the customers present in both the retail and loyalty databases.

#### 2.1.1.3 Application: Query Optimization with Set Operations
In database theory, set operations like union, intersection, and difference are used to optimize query processing. For instance, if you want to find the common records between two tables, you can use the intersection of sets. Lets use a more concrete example involving real-world data. Suppose we are working with two databases: one containing customer information from an online retail store and another from a loyalty program. We want to find customers who are both in the retail database and the loyalty program database.

***Application:*** Finding Common Customers in Retail and Loyalty Databases

***Concept:*** This application demonstrates how to find common customers across two datasets using set operations. This could be useful for targeted marketing or cross-promotions.


#### 2.1.1.4 Explanation:

**Retail Data:** Represents customers who have made purchases from an online retail store.

**Loyalty Data:** Represents customers who are members of a loyalty program.

**Set Conversion:** The customer records from both datasets are converted to sets of tuples, where each tuple represents a customer record.

**Set Operation:** The intersection of the two sets is computed to find customers who appear in both datasets.

**Output:** The result shows common customers, indicating those who are both retail customers and loyalty program members.

This practical example makes it clear how set operations can be applied to real-world data to find common entities across different datasets.

```python
# 2.1.1.6
import pandas as pd

# Sample data representing customer records from an online retail store
retail_data = {
    'CustomerID': [101, 102, 103, 104, 105], # Unique IDs for customers in the loyalty program
    'Name': ['John Smith', 'Jane Doe', 'Emily Johnson', 'Michael Brown', 'Sarah Davis'] # Corresponding customer 
}

# Sample data representing customer records from a loyalty program
loyalty_data = {
    'CustomerID': [104, 105, 106, 107, 108],
    'Name': ['Michael Brown', 'Sarah Davis', 'Laura Wilson', 'James Taylor', 'Olivia Anderson']
}

# Convert the sample data to DataFrames for easier manipulation
df_retail = pd.DataFrame(retail_data)
df_loyalty = pd.DataFrame(loyalty_data)

# Convert DataFrames to sets of tuples for set operations
# The 'itertuples' method is used to create tuples from DataFrame rows, excluding index and column names
set_retail = set(df_retail.itertuples(index=False, name=None))
set_loyalty = set(df_loyalty.itertuples(index=False, name=None))

# Find intersection of the two sets to identify common customers
common_customers = set_retail & set_loyalty
common_customers_df = pd.DataFrame(list(common_customers), columns=['CustomerID', 'Name'])

# Display the result
print("Common Customers in Retail and Loyalty Databases:")
print(common_customers_df)
```

### 2.1.2 Probability
Basic probability theory, includes events, sample spaces, and conditional probability (Ghahramani, 2024). This example demonstrates the application of conditional probability, a fundamental concept in probability theory, by calculating the likelihood of an event occurring given that another event has already occurred. The derivation and code snippet illustrate how to compute conditional probability using a practical scenario, such as determining the probability of a patient having a disease given a positive test result.

#### 2.1.2.1 Real Life Application of Probability

A real-world use case of conditional probability is in **medical testing**. Suppose a medical test is designed to detect a particular disease. Let:

- **Event $A$**: The patient has the disease.
- **Event $B$**: The test result is positive.

Given that a patient tests positive (Event $B$), doctors are interested in finding the probability that the patient actually has the disease (Event $A$). This is the conditional probability $P(A \mid B)$. Understanding this probability helps in making informed decisions about further testing, treatment, or monitoring.

#### 2.1.2.2 Explanation: Conditional Probability in Discrete Mathematics

In this example, we explore the concept of **conditional probability** using a simple scenario involving two coin flips. 

- **Sample Space**: We first define the sample space as all possible outcomes of two coin flips, which are $\{HH, HT, TH, TT\}$.
  
- **Event $A$**: "At least one head" is the event where the outcome contains at least one 'H'. The possible outcomes satisfying this condition are $\{HH, HT, TH\}$.

- **Event $B$**: "The first flip is a head" is the event where the first coin flip results in 'H'. The outcomes satisfying this condition are $\{HH, HT\}$.

- **Intersection of Events $A$ and $B$**: We then find the intersection of events $A$ and $B$, which is $\{HH, HT\}$, representing outcomes where both conditions are true.

- **Conditional Probability $P(A \mid B)$**: Finally, we calculate the conditional probability $P(A \mid B)$ using the formula:

  $P(A \mid B) = \frac{P(A \cap B)}{P(B)}$
  
  This gives us the probability of event $A$ ("at least one head") given that event $B$ ("the first flip is a head") has occurred.

The result is the probability that at least one head occurs given that the first flip is a head.


```python
# 2.1.2.2
import itertools

# Define the sample space for two coin flips (H = Heads, T = Tails)
sample_space = list(itertools.product(['H', 'T'], repeat=2))

# Define the event A: "At least one head"
event_A = [outcome for outcome in sample_space if 'H' in outcome]

# Define the event B: "The first flip is a head"
event_B = [outcome for outcome in sample_space if outcome[0] == 'H']

# Calculate P(A) = |A| / |Sample Space|
P_A = len(event_A) / len(sample_space)

# Calculate P(B) = |B| / |Sample Space|
P_B = len(event_B) / len(sample_space)

# Calculate P(A and B) = |A  B| / |Sample Space|
event_A_and_B = [outcome for outcome in event_A if outcome in event_B]
P_A_and_B = len(event_A_and_B) / len(sample_space)

# Calculate the conditional probability P(A | B) = P(A and B) / P(B)
P_A_given_B = P_A_and_B / P_B

# Output the result
print(f"P(A | B) = {P_A_given_B:.2f}")
```

### 2.1.3 Counting Principles
Counting priniciples include the Pigeonhole Principle, permutations, combinations, and the concept of factorials (Erciyes & Erciyes, 2021). In this example, we explore the application of counting principles in real-world scenarios, specifically focusing on organizing a project team and assigning tasks. We derive the formulas for combinations and permutations to calculate the number of ways to select team members and assign tasks, respectively, and then illustrate these concepts with a Python code snippet to solve the problem.

#### 2.1.3.1 Real-World Example: Organizing a Team and Selecting Projects

**Scenario:**

Imagine you are the manager of a team at a tech company. You have 5 employees and need to select 3 of them to form a project team, and then decide the order in which they will handle specific tasks. Additionally, there are 3 distinct projects, and you need to choose one project for each of the team members.

1. **Choosing Team Members (Combinations):**

   You want to select 3 employees out of the 5 available employees for the project team. The order in which you select the team members doesnt matter for this decision, so you use combinations.

   Using the formula for combinations:
   
   $\binom{5}{3} = \frac{5!}{3! \times (5-3)!} = 10$
   
   There are 10 different ways to choose 3 employees from the 5 available.

2. **Assigning Tasks (Permutations):**

   Once you have selected the 3 team members, you need to assign them to 3 distinct tasks. The order in which the tasks are assigned is important, so you use permutations.

   Using the formula for permutations:
   
   $P(3, 3) = \frac{3!}{(3-3)!} = 6$
   
   There are 6 different ways to assign the 3 tasks to the 3 selected team members.

**Summary:**

By using combinations, you determine the number of ways to form the team, and by using permutations, you determine the number of ways to assign tasks. This approach ensures you can effectively manage team selection and task assignments in a structured and organized manner.

#### 2.1.3.2 Derivation of Counting Principles

1. **Pigeonhole Principle**

   The Pigeonhole Principle states that if $n$ items are distributed into $k$ containers (or pigeonholes) and $n > k$, then at least one container must contain more than one item. Formally, if $n$ pigeons are placed in $k$ pigeonholes, where $n > k$, then:

   $\text{At least one pigeonhole contains } \lceil \frac{n}{k} \rceil \text{ or more pigeons.}$

2. **Permutations**

   Permutations refer to the number of ways to arrange $n$ distinct items. The number of permutations of $n$ distinct items is given by $n!$ (n factorial), which is defined as:

   $n! = n \times (n-1) \times (n-2) \times \cdots \times 1$

   For example, the number of permutations of 5 distinct items is $5!$:

   $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$

3. **Combinations**

   Combinations refer to the number of ways to choose $r$ items from $n$ distinct items without regard to the order of selection. The number of combinations is given by:

   $\binom{n}{r} = \frac{n!}{r! \times (n-r)!}$

   For example, the number of ways to choose 3 items from 5 distinct items is:

   $\binom{5}{3} = \frac{5!}{3! \times (5-3)!} = \frac{120}{6 \times 2} = 10$

4. **Example Using Permutations and Combinations**

   - To find the number of permutations of 3 items from a set of 5 items:

     $\text{Permutations of 3 items} = P(5, 3) = \frac{5!}{(5-3)!} = \frac{120}{2} = 60$

   - To find the number of combinations of 3 items from a set of 5 items:

     $\text{Combinations of 3 items} = \binom{5}{3} = 10$

```python
# 2.1.3.3 This code snippet demonstrates key counting principles in discrete mathematics. 
# The Pigeonhole Principle is used to assert that placing 10 pigeons in 9 pigeonholes guarantees at least one pigeonhole with more than one pigeon. 
# The code also calculates permutations and combinations, highlighting the difference between ordered arrangements and unordered selections. 
# Finally, the itertools library is used to generate and display permutations and combinations of a list of items.

from math import factorial
from itertools import permutations, combinations

# Pigeonhole Principle: If 10 pigeons are placed in 9 pigeonholes, at least one pigeonhole must contain more than one pigeon.
pigeons = 10
pigeonholes = 9
assert pigeons > pigeonholes, "By the Pigeonhole Principle, at least one pigeonhole must contain more than one pigeon."

# Permutations: Calculate the number of ways to arrange 5 distinct items
n = 5
perm = factorial(n)
print(f"The number of permutations of {n} distinct items is {perm}")

# Combinations: Calculate the number of ways to choose 3 items from 5 distinct items
r = 3
comb = factorial(n) // (factorial(r) * factorial(n - r))
print(f"The number of combinations of {n} distinct items taken {r} at a time is {comb}")

# Using itertools for permutations and combinations
items = ['A', 'B', 'C', 'D', 'E']

# Generate all permutations of 3 items
perm_list = list(permutations(items, r))
print(f"Permutations of 3 items from {items}: {perm_list}")

# Generate all combinations of 3 items
comb_list = list(combinations(items, r))
print(f"Combinations of 3 items from {items}: {comb_list}")
```

### 21 Functions and Relations
Functions and Relations explore how elements from one set map to another, including concepts like injective, surjective, and bijective functions (Boules, 2021). In this section, we analyze the normalization of a database relation using functional dependencies. We begin by defining the relation and its attributes, followed by specifying the functional dependencies that describe how certain attributes determine others. We then verify if the relation is in First Normal Form (1NF) by ensuring all attribute values are atomic. Next, we check for compliance with Second Normal Form (2NF) by ensuring that no non-prime attribute is partially dependent on any candidate key. The derivation confirms that the relation satisfies the conditions for both 1NF and 2NF, demonstrating the application of normalization principles in database design.

#### 2.2.1.1 Application: Functional Dependencies and Normalization in Database Design

**Concept:** Functional dependencies are used to ensure that database tables are properly normalized, reducing redundancy and improving data integrity. A relation is a table, and functional dependencies define how attributes in a relation are related to each other.

#### 2.2.1.2 Derivation: Functional Dependencies and Normalization

Given a relation $R$ with attributes and functional dependencies, we aim to determine if $R$ is in First Normal Form (1NF) and Second Normal Form (2NF). Consider a relation $R$ with the following attributes:

$R(\text{StudentID}, \text{CourseID}, \text{Instructor}, \text{Semester})$

***Functional Dependencies***

The functional dependencies for the relation are:

$$
\begin{aligned}
1. & \ \text{StudentID} \rightarrow \text{CourseID} \\
2. & \ \text{CourseID} \rightarrow \text{Instructor} \\
3. & \ \text{StudentID} \rightarrow \text{Semester}
\end{aligned}
$$

***First Normal Form (1NF)***

A relation is in 1NF if all its attributes contain only atomic (indivisible) values. 

$\text{Atomic values: } \text{StudentID}, \text{CourseID}, \text{Instructor}, \text{Semester}$

Since all attributes in $( R )$ contain atomic values, $( R )$ is in 1NF.

***Second Normal Form (2NF)***

A relation is in 2NF if:
1. It is in 1NF.
2. No non-prime attribute (an attribute that is not part of any candidate key) is partially dependent on any candidate key.

***Candidate Key***

Assume $( \text{StudentID} )$ is a candidate key.

***Checking Partial Dependencies***

To be in 2NF, there should be no partial dependency of non-prime attributes on a part of a candidate key. The non-prime attributes are:

$\text{CourseID}, \text{Instructor}, \text{Semester}$

***Checking dependencies:***

$$
\begin{aligned}
1. & \ \text{StudentID} \rightarrow \text{CourseID} \text{ (No partial dependency, as StudentID is the candidate key)} \\
2. & \ \text{CourseID} \rightarrow \text{Instructor} \text{ (Not a partial dependency on StudentID)} \\
3. & \ \text{StudentID} \rightarrow \text{Semester} \text{ (No partial dependency, as StudentID is the candidate key)}
\end{aligned}
$$

Since there are no partial dependencies of non-prime attributes on any candidate key part, $R$ is in 2NF.

*** Derivation Conclusion ***

Based on the derivation:
- The relation $R$ is in First Normal Form (1NF) because all attributes are atomic.
- The relation $R$ is in Second Normal Form (2NF) because there are no partial dependencies on any candidate key.

Thus, the relation satisfies the conditions for 1NF and 2NF, ensuring it is properly normalized.

#### 2.2.1.3 Python Code Example:

Heres a Python code snippet to illustrate the concept of functional dependencies and how to check if a relation is in First Normal Form (1NF) and Second Normal Form (2NF). The example involves checking functional dependencies in a relation and verifying normalization criteria.

#### 2.2.1.4 The Python code demonstrates how to:
1. **Define a Sample Relation:** A DataFrame `relation` represents a table with attributes like `StudentID`, `CourseID`, `Instructor`, and `Semester`.
2. **Define Functional Dependencies:** These are specified as tuples indicating how one attribute determines another.
3. **Check Normalization:**
   - **1NF Check:** Verifies that each cell contains atomic values, which means the relation is in 1NF.
   - **2NF Check:** Ensures that no non-prime attribute (attributes not part of any candidate key) is partially dependent on the candidate key. For simplicity, the code assumes `StudentID` as the candidate key.
By running the code, we determine whether the relation is in 1NF and 2NF, illustrating how functional dependencies and normalization principles are applied in database design.
This example showcases how discrete mathematics concepts are used in practical database schema design and optimization.

```python
# 2.2.1.5

import pandas as pd

# Define a sample relation with attributes and functional dependencies
relation = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'CourseID': ['CS101', 'CS102', 'CS101', 'CS103'],
    'Instructor': ['Dr. Smith', 'Dr. Jones', 'Dr. Smith', 'Dr. Brown'],
    'Semester': ['Fall', 'Spring', 'Fall', 'Summer']
})

# Define functional dependencies as tuples (determinant, dependent)
functional_dependencies = [
    ('StudentID', 'CourseID'),
    ('CourseID', 'Instructor'),
    ('StudentID', 'Semester')
]

def check_1nf(df):
    """Check if the relation is in First Normal Form (1NF)."""
    # 1NF requires that each cell contains atomic values (i.e., no lists or sets)
    for column in df.columns:
        if df[column].apply(lambda x: isinstance(x, (list, set, dict))).any():
            return False
    return True

def check_2nf(df, fds):
    """Check if the relation is in Second Normal Form (2NF)."""
    if not check_1nf(df):
        return False
    
    # 2NF requires that the relation is in 1NF and that no non-prime attribute
    # is partially dependent on any candidate key.
    # For simplicity, assume 'StudentID' is a candidate key here
    candidate_key = 'StudentID'
    
    partial_dependencies = []
    for determinant, dependent in fds:
        if determinant != candidate_key and dependent in df.columns:
            partial_dependencies.append((determinant, dependent))
    
    return len(partial_dependencies) == 0

# Check normalization
is_1nf = check_1nf(relation)
is_2nf = check_2nf(relation, functional_dependencies)

print("Is the relation in First Normal Form (1NF)?", is_1nf)
print("Is the relation in Second Normal Form (2NF)?", is_2nf)
```

#### 2.2.2.3 Time Complexity Derivation

1. **Linear Search**:
   - The algorithm iterates through the list exactly once to find the maximum value.
   - Time Complexity: $O(n)$, where $n$ is the length of the list.

2. **Quadratic Search**:
   - The algorithm compares each element with every other element in the list to determine if it's the maximum.
   - For each of the $n$ elements, it performs up to $n$ comparisons.
   - Time Complexity: $O(n^2)$, due to the nested loops.

#### 2.2.2.4 Comparison:
- The linear search is more efficient, especially for large lists, because its time complexity grows linearly with the input size, whereas the quadratic search's time complexity grows quadratically. This illustrates the practical importance of choosing algorithms with lower time complexity.

```python
# 2.2.2.4 
import time

# Linear search (O(n))
def linear_max(arr):
    max_val = arr[0]
    for num in arr:
        if num > max_val:
            max_val = num
    return max_val

# Quadratic search (O(n^2))
def quadratic_max(arr):
    n = len(arr)
    for i in range(n):
        is_max = True
        for j in range(n):
            if arr[j] > arr[i]:
                is_max = False
                break
        if is_max:
            return arr[i]

# Test with a large list
arr = list(range(10000))

# Time the linear search
start_time = time.time()
linear_max_val = linear_max(arr)
linear_time = time.time() - start_time

# Time the quadratic search
start_time = time.time()
quadratic_max_val = quadratic_max(arr)
quadratic_time = time.time() - start_time

print(f"Linear Search Time (O(n)): {linear_time:.5f} seconds")
print(f"Quadratic Search Time (O(n^2)): {quadratic_time:.5f} seconds")
```

This Python snippet illustrates the concept of **Big O notation** by comparing the time complexity of two different algorithms to find the maximum element in a list. The first algorithm, a linear search, has a time complexity of $O(n)$, meaning it scales linearly with the size of the input list. The second algorithm, a naive quadratic search, has a time complexity of $O(n^2)$, meaning it scales quadratically with the input size. The snippet times each algorithm and demonstrates how the quadratic algorithm is significantly slower as the list size increases, emphasizing the importance of understanding algorithmic efficiency in computational problems.

### 2.2.5.1 Real-World Example: Family Tree

```python
# 2.2.5.1
class TreeNode:
    def __init__(self, name, children=None):
        self.name = name
        self.children = children if children is not None else []

def print_tree(node, level=0):
    indent = ' ' * (level * 4)
    print(f"{indent}{node.name}")
    for child in node.children:
        print_tree(child, level + 1)

# Create the family tree
john = TreeNode('John', [
    TreeNode('Sarah', [
        TreeNode('Alice'),
        TreeNode('Bob')
    ]),
    TreeNode('David', [
        TreeNode('Michael')
    ])
])

# Print the family tree
print_tree(john)
```

#### 2.2.5.3 Derivation: Finding the Lowest Common Ancestor (LCA)

Given a binary tree and two nodes $p$ and $q$, we want to find their Lowest Common Ancestor (LCA). The LCA is defined as the deepest node that is an ancestor of both $p$ and $q$.

1. **Base Cases:**
   - If the root is $\text{None}$ (i.e., the tree is empty), then $\text{LCA}(root, p, q) = \text{None}$.
   - If the root is either $p$ or $q$, then the root itself is the LCA because the LCA must be one of the nodes if it matches either $p$ or $q$.

2. **Recursive Case:**
   - Recursively find the LCA in the left subtree: $\text{left} = \text{LCA}(root.left, p, q)$.
   - Recursively find the LCA in the right subtree: $\text{right} = \text{LCA}(root.right, p, q)$.

3. **Decision Making:**
   - If both $\text{left}$ and $\text{right}$ are non-null, then $root$ is the LCA because $p$ and $q$ are found in both subtrees.
   - If only one of $\text{left}$ or $\text{right}$ is non-null, then the LCA is in that non-null subtree.
   - If both are null, then $\text{LCA} = \text{None}$.

4. **Final Result:**
   - The LCA is returned as $root$ when $\text{left}$ and $\text{right}$ are both non-null, otherwise it is either $\text{left}$ or $\text{right}$, depending on which is non-null.

The recursive approach ensures that the LCA is found in $O(N)$ time complexity, where $N$ is the number of nodes in the tree.



```python
# 2.2.5.4 This code snippet finds the Lowest Common Ancestor (LCA) of two nodes in a binary tree using a recursive Depth-First Search (DFS) approach. 
# It traverses the tree to locate the nodes and their LCA efficiently by checking if both nodes are found in the left and right subtrees.
class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def lca(root, p, q):
    if not root or root == p or root == q:
        return root

    left = lca(root.left, p, q)
    right = lca(root.right, p, q)

    if left and right:
        return root
    return left if left else right

# Example usage
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

lca_node = lca(root, root.left.left, root.left.right)
print(f"The LCA of nodes 4 and 5 is: {lca_node.value}")
```

#### 2.2.6 Recurrence Relations
How sequences are defined using recurrence relations, and methods to solve them (Mazur, 2022). In this section, we explore a real-world application of recurrence relations in inventory management. 
Well derive the total inventory needed for a given month using a recurrence relation, and then illustrate this with a Python code snippet.

```python
# 2.2.6.1 Python Code to Compute and Visualize $T(n)$
import matplotlib.pyplot as plt

# Function to calculate T(n) using dynamic programming
def compute_T(n):
    # Base cases
    if n == 0:
        return 2
    elif n == 1:
        return 3
    
    # Initialize an array to store the computed values
    T = [0] * (n + 1)
    T[0] = 2
    T[1] = 3
    
    # Compute T(n) using the recurrence relation
    for i in range(2, n + 1):
        T[i] = 2 * T[i - 1] + 3 * T[i - 2]
    
    return T

# Define the number of terms to compute
n_terms = 5
T_values = compute_T(n_terms)

# Plotting the values of T(n)
plt.figure(figsize=(10, 6))
plt.bar(range(n_terms + 1), T_values, color='skyblue')
plt.xlabel('n')
plt.ylabel('T(n)')
plt.title('Visualization of T(n) using Dynamic Programming')
plt.xticks(range(n_terms + 1))
plt.grid(True)
plt.show()
```

#### 2.2.6.2 Real-Life Example: Inventory Management

Imagine a company managing the inventory for two products, A and B. The total inventory needed each month depends on the demand in the previous two months. Specifically, the demand in the next month is modeled by a recurrence relation where the inventory needed for month $n$ is twice the inventory from month $n-1$ plus three times the inventory from month $n-2$.

This can be represented as:

$T(n) = 2T(n-1) + 3T(n-2)$

Where $T(n)$ represents the total inventory for month $n$. This model helps the company predict future inventory needs based on past demand, ensuring they are prepared to meet customer demand.

#### 2.2.6.3 Recursive Derivation

Given the recurrence relation:

$T(n) = 2T(n-1) + 3T(n-2)$

with initial conditions:

$T(1) = 3, \quad T(0) = 2$

Let's calculate $T(5)$:

$$
\begin{aligned}
T(2) & = 2T(1) + 3T(0) = 2(3) + 3(2) = 6 + 6 = 12, \\
T(3) & = 2T(2) + 3T(1) = 2(12) + 3(3) = 24 + 9 = 33, \\
T(4) & = 2T(3) + 3T(2) = 2(33) + 3(12) = 66 + 36 = 102, \\
T(5) & = 2T(4) + 3T(3) = 2(102) + 3(33) = 204 + 99 = 303.
\end{aligned}
$$

Thus, $T(5) = 303$.

#### 2.2.6.4 Explanation

The goal is to find $T(n)$ using dynamic programming. The recurrence relation is a linear homogeneous recurrence relation of order 2, meaning that each term $T(n)$ is expressed as a linear combination of the previous two terms.

The steps to solve the problem are as follows:

1. **Base Cases**: 
   - For $n = 0$, we are given $T(0) = 2$.
   - For $n = 1$, we are given $T(1) = 3$.

2. **Recursive Formula**: 
   - For $n \geq 2$, we use the recurrence relation $T(n) = 2T(n-1) + 3T(n-2)$.


```python
# 2.2.6.2 Function to solve the recurrence relation using dynamic programming
def solve_recurrence(n):
    # Base cases
    if n == 0:
        return 2
    elif n == 1:
        return 3
    
    # Initialize the base cases
    T = [0] * (n + 1)
    T[0] = 2
    T[1] = 3
    
    # Use dynamic programming to fill the array
    for i in range(2, n + 1):
        T[i] = 2 * T[i-1] + 3 * T[i-2]
    
    return T[n]

# Example: Calculate T(5)
n = 5
result = solve_recurrence(n)
print(f"T({n}) = {result}")
```

#### .3..3.1.2 Derivation of Propositional Logic Operations

1. **Logical AND (Conjunction)**

   The logical AND operation, denoted as $p \land q$, is true if and only if both $p$ and $q$ are true. Mathematically, it can be expressed as:

   $$
   p \land q = 
   \begin{cases} 
   \text{True} & \text{if } p = \text{True} \text{ and } q = \text{True} \\
   \text{False} & \text{otherwise}
   \end{cases}
   $$

2. **Logical OR (Disjunction)**

   The logical OR operation, denoted as $p \lor q$, is true if at least one of $p$ or $q$ is true. Mathematically, it can be expressed as:

   $$
   p \lor q = 
   \begin{cases} 
   \text{True} & \text{if } p = \text{True} \text{ or } q = \text{True} \\
   \text{False} & \text{if } p = \text{False} \text{ and } q = \text{False}
   \end{cases}
   $$

3. **Logical NOT (Negation)**

   The logical NOT operation, denoted as $\neg p$, inverts the truth value of $p$. Mathematically, it can be expressed as:

   $$
   \neg p = 
   \begin{cases} 
   \text{True} & \text{if } p = \text{False} \\
   \text{False} & \text{if } p = \text{True}
   \end{cases}
   $$

4. **Logical Implication**

   The logical implication, denoted as $p \rightarrow q$, is true if $p$ is false or $q$ is true. Mathematically, it can be expressed as:
   
   $p \rightarrow q = \neg p \lor q$

   Which simplifies to:

   $$
   p \rightarrow q = 
   \begin{cases} 
   \text{True} & \text{if } p = \text{False} \text{ or } q = \text{True} \\
   \text{False} & \text{if } p = \text{True} \text{ and } q = \text{False}
   \end{cases}
   $$

```python
# 2.3.1.3
import pandas as pd
import itertools

# Define logical operators
def and_op(p, q):
    return p and q

def or_op(p, q):
    return p or q

def not_op(p):
    return not p

def implies(p, q):
    return not p or q

# Define truth values
truth_values = [True, False]

# Generate truth table
truth_table = []
for p, q in itertools.product(truth_values, repeat=2):
    row = {
        'p': p,
        'q': q,
        'p AND q': and_op(p, q),
        'p OR q': or_op(p, q),
        'NOT p': not_op(p),
        'p IMPLIES q': implies(p, q),
    }
    truth_table.append(row)

# Create DataFrame for better visualization
df_truth_table = pd.DataFrame(truth_table)

# Display the truth table
print(df_truth_table)
```

#### 2.3.2.3 Derivation

To simplify the Boolean expression:
$E = (A \land B \land \neg C) \lor (\neg A \land C) \lor (A \land \neg B \land C)$

we can apply Boolean algebra rules as follows:

1. **Combine Terms:**

   $E = (A \land B \land \neg C) \lor (A \land \neg B \land C) \lor (\neg A \land C)$

2. **Group Terms with Common Factors:**

   $E = [ (A \land B \land \neg C) \lor (A \land \neg B \land C) ] \lor (\neg A \land C)$

3. **Factor Out Common Terms:**

   $E = A \land [ (B \land \neg C) \lor (\neg B \land C) ] \lor (\neg A \land C)$

4. **Simplify Inner Expression:**
   
   $(B \land \neg C) \lor (\neg B \land C) = (B \oplus C)$

   where $( \oplus )$ denotes the XOR operation.

5. **Substitute Back:**

   $E = A \land (B \oplus C) \lor (\neg A \land C)$

   This is the simplified form of the original Boolean expression.

   Therefore, the simplified Boolean expression is:
   
   $E = A \land (B \oplus C) \lor (\neg A \land C)$

#### 2.3.3.2 Mathematical Derivation: DFA Recognizing Strings Ending in `01`

We define a Deterministic Finite Automaton (DFA) $( M = (Q, \Sigma, \delta, q_0, F) )$ to recognize strings over the alphabet $( \Sigma = \{0, 1\} )$ that end with the sequence `01`. The components of the DFA are as follows:

- **States**: $( Q = \{q_0, q_1, q_2\} )$
- **Alphabet**: $( \Sigma = \{0, 1\} )$
- **Transition Function**: $( \delta: Q \times \Sigma \rightarrow Q )$
- **Start State**: $( q_0 )$
- **Accepting State**: $( F = \{q_2\} )$

***Transition Function***

The transition function $( \delta )$ is defined as:
$$
\begin{align*}
[ \delta(q_0, 0) = q_1, \quad \delta(q_0, 1) = q_0 ] \\
[ \delta(q_1, 0) = q_1, \quad \delta(q_1, 1) = q_2 ] \\
[ \delta(q_2, 0) = q_1, \quad \delta(q_2, 1) = q_0 ]
\end{align*}
$$

***Explanation***

1. **Initial State**: The DFA starts in state $( q_0 )$.
2. **Reading `0`**: If the DFA reads a `0` while in state $( q_0 )$, it transitions to state $( q_1 )$.
3. **Reading `1` after `0`**: If the DFA reads a `1` while in state $( q_1 )$, it transitions to the accepting state $( q_2 )$, recognizing that the string ends in `01`.
4. **Reading Additional Symbols**: Any further symbols lead the DFA back to a non-accepting state unless the final sequence is `01`.

***Formal Description of Language***

The language $( L(M) )$ recognized by this DFA is:
$[ L(M) = \{w \in \{0, 1\}^* \mid \text{the last two symbols of } w \text{ are } 01\} ]$

***Verification***

To verify that the DFA correctly recognizes strings ending in `01`, consider an input string $( w = 1101 )$:

1. Start in $( q_0 )$.
2. Read `1`: Transition from $( q_0 )$ to $( q_0 )$.
3. Read `1`: Remain in $( q_0 )$.
4. Read `0`: Transition from $( q_0 )$ to $( q_1 )$.
5. Read `1`: Transition from $( q_1 )$ to $( q_2 )$.

Since the DFA ends in the accepting state $( q_2 )$, the string $( w = 1101 )$ is accepted, confirming that it ends in `01`.

***Conclusion***

This derivation illustrates how a DFA can be constructed to recognize specific patterns within strings. The DFA $( M )$ successfully identifies strings that end in the sequence `01`, demonstrating a fundamental concept in automata theory.

#### 2.3.3.4 Application

This Python code provides a practical implementation of automata theory by simulating a DFA. DFAs are essential in various computer science fields, including compiler design, formal language processing, and pattern recognition. By defining a DFA to recognize strings ending in `01`, this example illustrates how automata can be used to analyze and process sequences of symbols in computational systems.


```python
# 2.3.3.4 This code snippet demonstrates the concept of a **Deterministic Finite Automaton (DFA)**, a core topic in automata theory. 
# A DFA is defined by a set of states and transitions, where each transition is determined by the current state and an input symbol. 
# In this example, the DFA is constructed to recognize strings over the alphabet `{0, 1}` that end in the sequence `01`.
# - **DFA Definition**: The DFA is represented as a dictionary, where each key corresponds to a state, and the values represent transitions based on the input characters `'0'` and `'1'`.
# - **Initial and Accepting States**: The DFA starts in the initial state `q0` and attempts to reach the accepting state `q2`.
# - **DFA Simulation Function**: The `simulate_dfa` function processes an input string character by character, transitioning between states according to the DFA's rules.
# It returns `True` if the DFA ends in the accepting state, meaning the input string is recognized by the automaton.
# Define the DFA as a dictionary
dfa = {
    'q0': {'0': 'q1', '1': 'q0'},
    'q1': {'0': 'q1', '1': 'q2'},
    'q2': {'0': 'q1', '1': 'q0'}
}

# Set the initial and accepting states
initial_state = 'q0'
accepting_state = 'q2'

# Function to simulate the DFA
def simulate_dfa(dfa, initial_state, accepting_state, input_string):
    state = initial_state
    for char in input_string:
        state = dfa[state][char]
    return state == accepting_state

# Test the DFA with a string
input_string = "1101"
if simulate_dfa(dfa, initial_state, accepting_state, input_string):
    print(f"The string '{input_string}' is accepted by the DFA.")
else:
    print(f"The string '{input_string}' is not accepted by the DFA.")
```

   $\forall x \, P(x) \text{ means that } P(x) \text{ is true for all } x \text{ in the domain.}$

   For example, the statement:

   $\forall x \, (x > 0 \implies x^2 > 0)$

2. **Existential Quantifier**

   The existential quantifier $\exists$ (there exists) is used to assert that there is at least one element in the domain for which the predicate $P(x)$ is true. In logical terms:

   $\exists x \, P(x) \text{ means that there is at least one } x \text{ in the domain for which } P(x) \text{ is true.}$

   For example, the statement:

   $\exists x \, (x^2 = 4)$

   asserts that there exists some $x$ such that $x^2$ equals 4, which is true for $x = 2$ and $x = -2$.

3. **Combining Quantifiers**

   - **Universal Implication**: 

     $\forall x \, (P(x) \implies Q(x))$

     This states that for every $x$, if $P(x)$ is true, then $Q(x)$ must also be true.

   - **Existential Conjunction**: 

     $\exists x \, (P(x) \land Q(x))$

     This states that there exists an $x$ for which both $P(x)$ and $Q(x)$ are true.

### 2.5 Number Theory
Number theory covers divisibility, prime numbers, greatest common divisors, and modular arithmetic (De Koninck & Luca, 2023). In this example, we apply the Chinese Remainder Theorem to solve a scheduling problem where multiple recurring events must align on a common day. We will derive the solution using the theorem and implement it in Python to find a day that satisfies the constraints for all events.


```python
# 2.5.1 Visualization of Scheduling Problem Using CRT
import numpy as np
import matplotlib.pyplot as plt
from sympy import mod_inverse

# Function to solve the system of congruences using CRT
def chinese_remainder_theorem(a, n):
    N = np.prod(n)  # Product of all moduli
    x = 0
    for ai, ni in zip(a, n):
        Ni = N // ni
        Mi = mod_inverse(Ni, ni)
        x += ai * Mi * Ni
    return x % N

# Given data
a = [2, 3, 4]  # Remainders
n = [3, 5, 7]  # Moduli

# Solve the system of congruences
x = chinese_remainder_theorem(a, n)
N = np.prod(n)  # The period where all events coincide

# Visualization
days = np.arange(1, N + 1)
events_A = (days % n[0] == a[0])
events_B = (days % n[1] == a[1])
events_C = (days % n[2] == a[2])
coincidence = events_A & events_B & events_C

plt.figure(figsize=(12, 6))
plt.plot(days, events_A, label="Event A (Every 3 days)", marker='o', linestyle='None')
plt.plot(days, events_B, label="Event B (Every 5 days)", marker='x', linestyle='None')
plt.plot(days, events_C, label="Event C (Every 7 days)", marker='s', linestyle='None')
plt.plot(days, coincidence, label="Coincidence", color='red', marker='D', linestyle='None')
plt.axvline(x, color='black', linestyle='--', label=f"First Coincidence at day {x}")
plt.xlabel('Days')
plt.ylabel('Event Occurrence')
plt.title('Visualization of Scheduling Problem Using CRT')
plt.legend()
plt.grid(True)
plt.show()

print(f"The first common day for all events is day {x}.")
print(f"All events will coincide every {N} days.")
```

#### 2.5.1.2 Practical Application: Scheduling and Time Management

Imagine a scheduling problem where you need to find a common time slot that fits multiple recurring events. For instance:

$$

you can find a common day $x$ when all events coincide. This helps in scheduling tasks, optimizing resource allocation, and managing overlapping activities efficiently.

#### 2.5.1.3 Derivation of the Solution Using the Chinese Remainder Theorem

To solve the system of congruences:

$$
\begin{aligned}
x &\equiv 2 \pmod{3} \\
x &\equiv 3 \pmod{5} \\
x &\equiv 4 \pmod{7}
\end{aligned}
$$

we use the Chinese Remainder Theorem. The theorem states that if the moduli are pairwise coprime, there is a unique solution modulo the product of the moduli.

1. **Combine the Congruences**

   We start by solving the first two congruences:

   $$
   \begin{aligned}
   x &\equiv 2 \pmod{3} \\
   x &\equiv 3 \pmod{5}
   \end{aligned}
   $$

   Let $x = 3k + 2$. Substituting this into the second congruence:

   $3k + 2 \equiv 3 \pmod{5}$

   Solving for $k$:

   $3k \equiv 1 \pmod{5}$

   The multiplicative inverse of 3 modulo 5 is 2 (since $3 \times 2 \equiv 1 \pmod{5}$). Thus:

   $k \equiv 2 \pmod{5}$

   So, $k = 5m + 2$. Substituting $k$ back:

   $x = 3(5m + 2) + 2 = 15m + 8$

   Thus:

   $x \equiv 8 \pmod{15}$

2. **Combine with the Third Congruence**

   Now, use the result from above to solve:

   $$
   \begin{aligned}
   x &\equiv 8 \pmod{15} \\
   x &\equiv 4 \pmod{7}
   \end{aligned}
   $$

   Let $x = 15n + 8$. Substituting into the third congruence:

   $15n + 8 \equiv 4 \pmod{7}$

   Simplify:

   $15n \equiv -4 \pmod{7}$

   Since $15 \equiv 1 \pmod{7}$:

   $n \equiv -4 \pmod{7}$

   The solution is:

   $n \equiv 3 \pmod{7}$

   Therefore, $n = 7p + 3$. Substituting back:

   $x = 15(7p + 3) + 8 = 105p + 53$

   Thus:
   
   $x \equiv 53 \pmod{105}$

   The smallest non-negative solution is $x = 53$. 

So, the solution to the system of congruences is $x \equiv 53 \pmod{105}$.


#### 2.5.1.3 Explanation of the Chinese Remainder Theorem Implementation

1. **Extended Euclidean Algorithm (`extended_gcd`)**:
   - The Extended Euclidean Algorithm is used to compute the greatest common divisor (GCD) of two integers \(a\) and \(b\). Additionally, it finds integer coefficients \(x\) and \(y\) such that:
     $$
     \text{gcd}(a, b) = ax + by
     $$
   - This is useful for solving linear Diophantine equations and for finding modular inverses.

2. **Chinese Remainder Theorem (`chinese_remainder_theorem`)**:
   - The Chinese Remainder Theorem (CRT) solves a pair of congruences:
     $$
     x \equiv a_1 \pmod{n_1}
     $$
     $$
     x \equiv a_2 \pmod{n_2}
     $$
   - Using the CRT, we combine these congruences into a single congruence modulo the least common multiple (LCM) of \(n_1\) and \(n_2\). The solution is found by solving:
     $$
     x = (a_1 + (a_2 - a_1) \cdot m_1 \cdot n_1) \mod \text{lcm}(n_1, n_2)
     $$
   - where \(m_1\) is the coefficient found using the Extended Euclidean Algorithm.

3. **Solving Multiple Congruences (`solve_congruences`)**:
   - To solve a system of multiple congruences, we iteratively apply the Chinese Remainder Theorem to pairs of congruences. This reduces the problem step-by-step until a single congruence represents the solution to the entire system.

This approach ensures that we find an \(x\) which satisfies all the given congruences simultaneously by leveraging modular arithmetic and the properties of the Chinese Remainder Theorem.


```python
# 2.5.1.3
def extended_gcd(a, b):
    """Extended Euclidean Algorithm to find the gcd of a and b."""
    if a == 0:
        return (b, 0, 1)
    gcd, x1, y1 = extended_gcd(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return (gcd, x, y)

def chinese_remainder_theorem(a1, n1, a2, n2):
    """Solves the system of two congruences using the Chinese Remainder Theorem."""
    gcd, m1, m2 = extended_gcd(n1, n2)
    if (a2 - a1) % gcd != 0:
        return None  # No solution
    lcm_n1_n2 = n1 * n2 // gcd
    x = (a1 + (a2 - a1) * m1 * n1) % lcm_n1_n2
    return (x, lcm_n1_n2)

def solve_congruences(congruences):
    """Solves a system of congruences using the Chinese Remainder Theorem."""
    x, mod = congruences[0]
    for a, n in congruences[1:]:
        x, mod = chinese_remainder_theorem(x, mod, a, n)
        if x is None:
            return None
    return (x % mod, mod)

# Define the congruences as (remainder, modulus)
congruences = [
    (2, 3),  # x  2 (mod 3)
    (3, 5),  # x  3 (mod 5)
    (4, 7)   # x  4 (mod 7)
]

# Solve the system of congruences
solution = solve_congruences(congruences)

# Print the solution
if solution:
    print(f"The solution to the system of congruences is x  {solution[0]} (mod {solution[1]})")
else:
    print("No solution found")
```
